```python
import pandas as pd
import seaborn as sns


# Assuming 'Churn' is the column indicating customer churn (1 for churn, 0 for not churn)
targetcolumn = 'price'

# Selecting categorical columns excluding the 'Churn' column
categorical_features = [column for column in df.select_dtypes(include='object').columns if column != targetcolumn]




# //////////
categorical_features = []

# List to store correlation coefficients
correlation_coefficients = []


for column in categorical_columns:
    if column != targetcolumn:
        categorical_features.append(column)
        correlation = df[column].astype('category').cat.codes.corr(df[targetcolumn])
        correlation_coefficients.append(correlation)
        print(f"The correlation between {column} and {targetcolumn} is: {correlation}")

# Sort the results in ascending order based on correlation coefficients
sorted_correlations = sorted(zip(categorical_features, correlation_coefficients), key=lambda x: x[1])

# Print the sorted results
print("\nCorrelation coefficients in ascending order:")
for feature, correlation in sorted_correlations:
    print(f"{feature}: {correlation}")
```

**Data Preparation for Modelling**

```python
import pandas as pd

# Assuming 'df' is your DataFrame
# If 'df' is not defined, replace it with the actual variable name of your DataFrame

# Select only columns with data type 'object'
categorical_columns = df.select_dtypes(include=['object']).columns

# Display the list of categorical columns
print("Categorical Columns:")
print(categorical_columns)
```

```python
column_to_drop = 'CarName'
df = df.drop(column_to_drop, axis=1)

# Display the DataFrame after dropping the column
print(df.head())
```

```python
#one hot encoding for all the categorical columns to be able to make linear regression

categorical_columns = ['fueltype', 'aspiration', 'doornumber', 'carbody','drivewheel',
       'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']
df_encode = pd.get_dummies(data=df,
                           columns=categorical_columns,
                           drop_first=True,
                           dtype='int8')

# Let's have a look at the difference in data before/after encoding
print('Columns in original data frame:\n',df.columns.values)
print('\nNumber of rows and columns in the dataset:',df.shape)
print('\nColumns in data frame after encoding dummy variable:\n',df_encode.columns.values)
print('\nNumber of rows and columns in the dataset:',df_encode.shape)
```

**Modelling**

```python
from sklearn.model_selection import train_test_split

x = df_encode.drop('price',axis=1) # Independent variable
y = df_encode['price'] # Dependent variable

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=23)

print('Training dataset size:', x_train.shape[0])
print('Testing dataset size:', x_test.shape[0])
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train);
# sklearn regression module
y_pred = lin_reg.predict(x_test)









```

**Evaluation**

```python
# Evaluation
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_pred, y_test, squared=False)
```

```python
from sklearn.metrics import r2_score


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error (MSE) is: {mse}")
print(f"R-squared (R²) is: {r2}")
```
